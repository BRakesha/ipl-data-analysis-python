```python
teams_played=pd.concat([ipl['team1'],ipl['team2']]).value_counts()
teams_wins=ipl['winner'].value_counts()

win_ratio={}
for team in pd.concat([ipl['team1'],ipl['team2']]).unique():
  win_ratio[team]= teams_wins[team]/teams_played[team]*100
```

## Which city has the highest average result margin?


```python
runwise_result_margin=ipl[(ipl['result']=='runs') & (~ipl['result'].isin(['tie','no result']))].groupby('city')['result_margin'].mean()

wicketwise_result_margin=ipl[(ipl['result']=='wickets') & (~ipl['result'].isin(['tie','no result']))].groupby('city')['result_margin'].mean()



print(f'{runwise_result_margin.idxmax()} Is Highest Result Margin Based On Runs {runwise_result_margin.max()}')

print(f'{wicketwise_result_margin.idxmax()} Is Highest Result Margin Based On wickets {wicketwise_result_margin.max()}')
```
